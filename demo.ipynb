```python
#导入数据
import pandas as pd
import numpy as np
df=pd.read_excel("E:\Huawei cloud\华为云盘\文档\毕业论文\stata代码\data.xlsx")
df.head()
```

```python
df['Vc'] = df['Vc'].astype('bool')
df.info()
```

```python
#删除空值
df = df.dropna()
df.info()
```

```python
#对数变换专利
df['Lnpatent'] = np.log(df['Patent'])
df = df.drop(['Patent'], axis=1)
df.info()
```

```python
df.describe()
```

```python
#正态标准化
from sklearn.preprocessing import StandardScaler
df_r = df.drop(['Stkcd', 'Year', 'Vc', 'IND'], axis=1)
scaler = StandardScaler().fit(df_r)
df[['Caperc','Lev','PPEperc','NPGR','SGR','RTR','ROA','Size','Lnpatent']] = scaler.transform(df_r)
df.describe()
```

```python
#数据平衡性
df['Vc'].value_counts()
```

```python
#将控制组取出
is_control = df.groupby('Stkcd')['Vc'].transform(lambda x: (x==0).all())
data_control = df[is_control]
data_control.info()
```

```python
df_sorted = df.sort_values(by=['Stkcd', 'Year'])
df['is_bad'] = df_sorted.groupby('Stkcd')['Vc'].transform('first') == 1
data_bad = df[df['is_bad']].copy()
data_bad.info()
```

```python
data = df.drop(data_bad.index)
data.info()
```

```python
data_treat = data.drop(data_control.index)
data_treat.info()
```

```python
#分别导出控制组和实验组
data_control.to_excel('E:\Huawei cloud\华为云盘\文档\毕业论文\stata代码\data_control.xlsx')
data_treat.to_excel('E:\Huawei cloud\华为云盘\文档\毕业论文\stata代码\data_treat.xlsx')
data_bad.to_excel('E:\Huawei cloud\华为云盘\文档\毕业论文\stata代码\data_bad.xlsx')
```
