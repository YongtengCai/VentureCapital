```python
#导入数据
import pandas as pd
import numpy as np
df=pd.read_excel("E:\Huawei cloud\华为云盘\文档\毕业论文\stata代码\data.xlsx")
df.head()
```

```python
df.info()
```

```python
#行业分类编码
df = df.join(pd.get_dummies(df['IND'], prefix='IND', drop_first=True))
df = df.drop(['IND'], axis=1)
df.info()
```

```python
#删除空值
df = df.dropna()
df.info()
```

```python
#对数变换专利
df['Lnpatent'] = np.log(df['Patent'])
df = df.drop(['Patent'], axis=1)
df.info()
```
